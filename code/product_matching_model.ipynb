```python
X_train.dropna(inplace=True)
```

**Import Model**

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier, Perceptron
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import VotingClassifier
```

```python
#define model
log_reg = LogisticRegression()
SGD = SGDClassifier()
percep = Perceptron()
tree = DecisionTreeClassifier()
rf = RandomForestClassifier()
knn = KNeighborsClassifier()
SVC = SVC()
lin_SVC = LinearSVC()
gauss = GaussianNB()
```

```python
tree.fit(X_train, y_train)
tree.score(X_train,y_train)
```

```python
#run the model
classifiers = [('Decision Tree Classifier', tree),
               ('Random Forest Classifier', rf),
              ('KNN', knn),
              ('Logistic Regression', log_reg),
              ('Perceptron', percep),
              ('Stochastic Gradient', SGD),
              ('SVC', SVC),
              ('Linear SVC', lin_SVC),
              ('GaussianNB', gauss)]
# iterate over the defined list of tuples containing the classifiers
for clf_name, clf in classifiers:
    # fit clf to the data
    clf.fit(X_train, y_train)
    
    # predict the labels of the test
    y_pred = clf.predict(X_train)
    
    # evaluate the accuracy of clf on the train set
    print('{:s} : {:.9f}'.format(clf_name, accuracy_score(y_train, y_pred)))
```

## Hyperparameter Tuning

```python
#import GridSearchCV
from sklearn.model_selection import GridSearchCV
```

**Decision Tree Classifier**

```python
#Decision Tree hyperparameter tuning
params_tree = {'max_depth': range(1,30),
               'min_samples_leaf':range(1,11),
               'min_samples_split': range(1,7),
              'criterion': ['entropy', 'gini']
              }
grid_tree = GridSearchCV(estimator = tree,
                        param_grid = params_tree,
                        scoring= 'accuracy',
                        n_jobs=-1,
                        )
grid_tree.fit(X_train, y_train)
```

```python
# get best parameter and score
print(grid_tree.best_params_)
print(grid_tree.best_score_)
print(grid_tree.best_estimator_)
```

**Random Forest Classifier **

```python
#Random Forest Hyperparameter Tuning
rf=RandomForestClassifier(oob_score=True, random_state=1)

params_rf = {'criterion': ["gini","entropy"], 
             'min_samples_leaf': [3,4,5], 
             'n_estimators': [50,100], 
             'min_samples_leaf':[2,4,10],
             'max_depth':[3,4,5,6,7],
             'max_features':['sqrt','log2']}

grid_rf = GridSearchCV(estimator = rf,
                        param_grid = params_rf,
                        scoring= 'accuracy',
                        n_jobs=-1
                      )
grid_rf.fit(X_train,y_train)
```

```python
print(grid_rf.best_params_)
print(grid_rf.best_score_)
print(grid_rf.best_estimator_)
```

```python
pd.DataFrame({'index':list(range(len(test))), 
                           'Label': grid_rf.best_estimator_.predict(X_test).astype(int)})
```

**KNN**

```python
#KNN hyperparameter tuning
param_grid_knn = {'leaf_size':range(1,15), 'n_neighbors': range(1,15),
                  'p':[1,2,5], 'weights':['uniform', 'distance'],
                 'algorithm': ['auto']}
grid_knn = GridSearchCV(estimator=knn, param_grid=param_grid_knn, scoring='accuracy',
                       n_jobs=-1)
grid_knn.fit(X_train, y_train)
```

```python
print(grid_knn.best_params_)
print(grid_knn.best_score_)
print(grid_knn.best_estimator_)
```

```python
pd.DataFrame({'index':list(range(len(test))), 
                           'Label': grid_knn.best_estimator_.predict(X_test).astype(int)})
```
